<h3> Student's t-test:</h3>

This is a independent t-test. <br/>
The hypothesis for this t-test is, <br/>
H0 (Null Hypothesis) = There is no statistical difference between two samples (both samples have identical averages).<br/>
H1 (Alternate Hypothesis) = The samples are statistically different. <br/><br/>
We will run our T-test based on 1000 iteraton samples of Total score, since it is the most appropriate attribute to measure distinguishability of game playing agents. <br/>
The test will be between -<br/>
1) Agent1 & Random agent <br/>
2) Agent2 & Random agent <br/>
3) Agent1 & Agent2

Test 1:

```python
ttest1 = ttest_ind(myA1.TotalScore, randA.TotalScore)
print("Test 1 results:")
print("t-value = {}".format(ttest1[0]))
print("p-value = {}".format(ttest1[1]))
```

As <b>t-value > p-value</b>,<br/>
We will reject our Null Hypothesis.<br/>
Hence, the two samples, Agent1 and Random agent are statistically different.

Test 2:

```python
ttest2 = ttest_ind(myA2.TotalScore, randA.TotalScore)
print("Test 2 results:")
print("t-value = {}".format(ttest2[0]))
print("p-value = {}".format(ttest2[1]))
```

As <b>t-value > p-value</b>,<br/>
We will reject our Null Hypothesis.<br/>
Hence, the two samples, Agent2 and Random agent are statistically different.

Test 3:

```python
ttest3 = ttest_ind(myA1.TotalScore, myA2.TotalScore)
print("Test 3 results:")
print("t-value = {}".format(ttest3[0]))
print("p-value = {}".format(ttest3[1]))
```

As <b>t-value > p-value</b>,<br/>
We will reject our Null Hypothesis.<br/>
Hence, the two samples, Agent1 and Agent2 are statistically different.

Thus we can conclude that, all the samples are statistically different.
